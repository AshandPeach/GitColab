```python
import pandas as pd
data = pd.read_csv(r"/content/drive/MyDrive/ASSISTments2017.csv",low_memory=False) 
#dtype={'Ln': 'float', 'Ln-1': 'str'}
#auto-generate index
#warning: Ln and Ln-1 have dtype('0')
data.head()
```

```python
data.info()
# 2   InferredGender  769160 non-null 
# 58   sumTime3SDWhen3RowRight  942731 non-null
# 81  isSTEM  316974 non-null
# features having incomplete values
```

```python
data.dropna(axis=0,inplace=True) #delete all rows have empty values
```

```python
Xdata = data
Xdata.drop(["correct"],axis=1,inplace=True)
```

```python
Xdata.head()
#Xdata.info()
```

```python
X = Xdata.iloc[:,0:81]
y = data.iloc[:,23] 
# X.shape
# X.head()
# y
```

```python
X = X.reset_index(drop=True)
X
```

```python
X.info()
#2   InferredGender
#3   SY ASSISTments Usage4
#15  skill
#17  problemType
#need to process to one-hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
GenderData = X.iloc[:,2]
SYData = X.iloc[:,3]
skillData = X.iloc[:,15]
problemTypeData = X.iloc[:,17]
noOfencFeature=4
```

```python
Tobeenc = pd.concat([GenderData,pd.DataFrame(SYData)],axis=1)
Tobeenc = pd.concat([Tobeenc,pd.DataFrame(skillData)],axis=1)
Tobeenc = pd.concat([Tobeenc,pd.DataFrame(problemTypeData)],axis=1)
Tobeenc
Tobeenc.info()
```

```python
enc = OneHotEncoder(categories='auto').fit(Tobeenc) #fit
```

```python
enc.transform(Tobeenc)
result = enc.transform(Tobeenc).toarray() #toarray
result
```

```python
#onehot-feature names

enc.get_feature_names_out()
```

```python
featureNames = enc.get_feature_names_out()
```

```python
X_ = pd.concat([X,pd.DataFrame(result)],axis=1)
X_.head()
```

```python
X_.drop(["InferredGender","SY ASSISTments Usage","skill","problemType"],axis=1,inplace=True) #axis=1代表删除列，inplace=True代表覆盖原数据
```

```python
X_.head()
```

```python
print(X_.columns)
```

```python
X.shape #(263750, 81)
noOfencFeature #4
X_.shape #(263750, 190)
X_.shape[1] #190
featureNames.shape[0] #113
X_.columns[76] #isSTEM
```

```python
# X.shape[1]-noOfencFeature #77
i=0
j=0
while i < featureNames.shape[0]:
  X_.rename(columns={i:featureNames[j]},inplace=True)
  i+=1
  j+=1
j
```

```python
X_
```

```python
#mutual information 
from sklearn.feature_selection import mutual_info_classif as MIC
result = MIC(X_,y)
result
```

```python
sum(result)
```

```python
(result == 0).sum() #38 non-related features
```

```python
FeatureIndex = pd.DataFrame(X_.columns)
FeatureIndex
```

```python
weightTable = pd.concat([FeatureIndex,pd.DataFrame(result)],axis=1)
weightTable.columns =["Feature","MutualInformation"]
weightTable
```
